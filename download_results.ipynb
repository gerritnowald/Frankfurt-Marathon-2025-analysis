```python
import requests
```

```python
for sex in ['M', 'W']:
    max_pages = 10 if sex == 'M' else 3
    for page in range(1, max_pages + 1):

        url = f"https://live.frankfurt-marathon.com/2023/?page={page}&event=L_HCH3BKLB3B8&num_results=1000&pid=list&search%5Bsex%5D={sex}&search%5Bage_class%5D=%25"

        # Send request with headers to mimic browser
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
        }

        response = requests.get(url, headers=headers)

        # download the page to html
        with open(f'results_{sex}_{page}.html', 'w', encoding='utf-8') as file:
            file.write(response.text)
```
